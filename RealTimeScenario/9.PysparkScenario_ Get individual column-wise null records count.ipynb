```
df = spark.read.csv(path='dbfs:/FileStore/data/empCompanyData.csv', header=True, inferSchema=True, nullValue='null')
df.show()
```

all columns null count

```
for i in df.columns:
    print(i, df.filter(f'{i} is null').count())
```

```
from pyspark.sql.functions import col, when, count
```

```
display(df.select([count(when(col(i).isNull(), i)).alias(i) for i in df.columns]))
```
