```
from pyspark.sql.functions import spark_partition_id
```

```
df = spark.read.csv(path='dbfs:/FileStore/data/empCompanyData.csv', header=True, inferSchema=True, nullValue='null')\
.repartition(4,'DEPTNO').withColumn('PartitionID', spark_partition_id())
display(df)
```

```
df.createOrReplaceTempView('df')
```

```
display(df.orderBy('SAL'))
```

```
display(df.sort('SAL'))
```

```
display(df.sortWithinPartitions('SAL'))
```

```
%sql
select * from df order by sal
```

```
%sql
select * from df sort by sal
```
