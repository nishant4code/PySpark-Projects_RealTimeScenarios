```
df1=spark.read.csv(path='dbfs:/FileStore/data/empCompanyData.csv', header=True, nullValue='null')
display(df1)
df1.printSchema()
```

```
from pyspark.sql.functions import to_date, col, date_format
```

```
df1.withColumn('HIREDATE', df1.HIREDATE).fillna('9999-12-31').show()
```

```
df2=df1.withColumn('HIREDATE', to_date('HIREDATE', 'dd-MM-yyyy')).fillna({'HIREDATE':'9999-12-31'})
df2.show()
```

```
#creating new columns as YEAR, MONTH from HIREDATE column
df3=df2.withColumn('YEAR', date_format('HIREDATE', 'yyyy')).withColumn('MONTH', date_format('HIREDATE', 'MM'))
df3.show()
```

```
df3.write.format('delta').partitionBy('YEAR', 'MONTH').mode('overwrite').saveAsTable('empDataPartitionByYearMonth')
```
