####difference between coalesce & repartition

```
rdd = sc.parallelize(range(10),4)
```

```
rdd.getNumPartitions()
```

```
#display data as per partition
rdd.glom().collect()
```

```
rdd.collect()
```

```
#let's reduce the number of partitions- both coalesce & repartition is used
rdd1 = rdd.coalesce(2)
rdd2 = rdd.repartition(2)
```

```
print('original RDD', rdd.glom().collect())
print('coalesce RDD', rdd1.glom().collect())
print('repartition RDD', rdd2.glom().collect())
```

```
rdd3 = rdd.coalesce(6) #doesnot work for increasing partition
rdd4 = rdd.repartition(6)
```

```
print('original RDD', rdd.glom().collect())
print('coalesce RDD', rdd3.glom().collect())
print('repartition RDD', rdd4.glom().collect())
```

```
rdd3 = rdd.coalesce(6, True)
rdd4 = rdd.repartition(6)
```

```
print('original RDD', rdd.glom().collect())
print('coalesce RDD', rdd3.glom().collect())
print('repartition RDD', rdd4.glom().collect())
```

```
df = spark.read.csv(path='dbfs:/FileStore/data/empCompanyData.csv', header=True)
display(df)
```

```
from pyspark.sql.functions import spark_partition_id
```

```
df1 = df.repartition(4).withColumn('partitionID', spark_partition_id())
display(df1)
```

```
df1.groupBy('partitionID').count().show()
```

```
#coalesce() can not increase partitions
df1 = df.coalesce(4).withColumn('partitionID', spark_partition_id())
display(df1)
```

```
#coalesce() can not increase partitions in dataframe, if tried returns error
df1 = df.coalesce(4, True).withColumn('partitionID', spark_partition_id())
display(df1)
```

```
df1 = df.rdd.coalesce(4, True).toDF().withColumn('partitionID', spark_partition_id())
display(df1)
```

```
df1.rdd.getNumPartitions()
```

```
df1.rdd.glom().collect()
```
