#how to skip first few lines while reading a csv

```
dbutils.fs.put('dbfs:/FileStore/spark_RealTimeScenarioData/skiplines.csv',"""line1
line2
line3
line4
id, name, loc
1, ravi, banglore
2, nitin, chennai
3, nishant, delhi
4, aakash, noida
""")
```

```
df=spark.read.csv(path='dbfs:/FileStore/spark_RealTimeScenarioData/skiplines.csv', header=True)
display(df)
```

```
rdd = sc.textFile('/FileStore/spark_RealTimeScenarioData/skiplines.csv')
rdd.collect()
```

```
rdd.zipWithIndex().collect()
```

```
rdd.zipWithIndex().filter(lambda x: x[1]>3).collect()
```

```
rdd.zipWithIndex().filter(lambda x: x[1]>3).map(lambda x:x[0]).collect()
```

```
rdd.zipWithIndex().filter(lambda x: x[1]>3).map(lambda x:x[0].split(',')).collect()
```

```
finalRDD = rdd.zipWithIndex().filter(lambda x: x[1]>3).map(lambda x:x[0].split(','))
columns = finalRDD.collect()[0]
columns
```

```
skipLine = finalRDD.first()
df=finalRDD.filter(lambda x: x!=skipLine).toDF(columns)
display(df)
```
